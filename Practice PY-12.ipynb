```python
def lucky_ticket(ticket_number):
    return (
       ticket_number//100000
     + ticket_number//10000 % 10
     + ticket_number//1000 % 10 ==
       ticket_number//100 % 10
     + ticket_number//10 % 10
     + ticket_number%10
)
    
print(lucky_ticket(523406))
```

```python
ratings_movies = pd.read_csv('data/ratings_movies.csv')
ratings_movies
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
# Создайте в таблице новый признак year_release, который соответствует году 
# выпуска фильма. У скольких фильмов не указан год их выпуска?
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
ratings_movies['year_release'].isnull().sum()
```

```python
# Какой фильм, выпущенный в 1999 году, получил наименьшую среднюю оценку зрителей?
rat_1999 = ratings_movies[ratings_movies['year_release'] == 1999]
rat_1999.groupby('title')['rating'].mean().sort_values().head(1)
```

```python
#Какое сочетание жанров фильмов (genres), выпущенных в 2010 году, получило наименьшую среднюю оценку (rating)?
rat_2010 = ratings_movies[ratings_movies['year_release'] == 2010]
rat_2010.groupby('genres')['rating'].mean().sort_values().head(1)
```

```python
# Какой пользователь (userId) посмотрел наибольшее количество различных
# (уникальных) комбинаций жанров (genres) фильмов?
ratings_movies.groupby('userId')['genres'].nunique().sort_values(ascending=False).head(1).index
```

```python
# Найдите пользователя, который выставил наименьшее количество оценок,
# но его средняя оценка фильмам наибольшая.
ratings_movies.groupby('userId')['rating'].agg(['count', 'mean']).sort_values(by=['count', 'mean'], ascending=[True, False])
```

```python
#Найдите сочетание жанров (genres) за 2018 году, которое имеет наибольший
# средний рейтинг (среднее по столбцу rating), и при этом число выставленных
# ему оценок (количество значений в столбце rating) больше 10.
rat_2018 = ratings_movies[ratings_movies['year_release'] == 2018]
rat_2018_mc = rat_2018.groupby('genres')['rating'].agg(['mean', 'count'])
rat_2018_mc[rat_2018_mc['count'] > 10].sort_values('mean').tail(1)

# .sort_values(by=['mean', 'count'], ascending=[False, False])
```

```python
ratings_movies['date'] = pd.to_datetime(ratings_movies['date'])
ratings_movies['year_rating'] = ratings_movies['date'].dt.year
pivot = ratings_movies.pivot_table(
    index='year_rating',
    columns='genres',
    values='rating',
    aggfunc='mean'
)
display(pivot[pivot.index == 2018]['Animation|Children|Mystery'])
```

```python
# Создайте сводную таблицу, которая иллюстрирует зависимость
# среднего рейтинга фильма от года выставления оценки и жанра.

rat_2018 = ratings_movies[ratings_movies['year_release'] == 2018]
rat_yg = rat_2018.groupby(['genres', 'year_release'])['rating'].max().unstack()
rat_yg[rat_yg[2018] == 5]
# # rat_yg['Comedy'].tail(23)
# rat_yg[rat_yg['rating'] == 5]
```

```python
orders = pd.read_csv('data/orders.csv', sep=';')
products = pd.read_csv('data/products.csv', sep=';')
orders
# products.shape
```

```python
# Объедините заданные таблицы в таблицу orders_products, чтобы в 
# результирующей таблице оказалась информация обо всех заказах, но 
# не оказалось информации о продуктах, на которых заказов ещё не поступало. 
orders_products = orders.merge(
    products,
    right_on='Product_ID',
    left_on='ID товара',
    how='left'
)

```

```python
# Какой идентификатор (Order ID) имеет заказ, для которого не оказалось информации о товаре?
orders_products[orders_products['Name'].isnull()]['Order ID']
```

```python
# На какой товар была произведена отмена?
orders_products[orders_products['Отменен'] == 'Да']
```

```python
# Какой покупатель принёс наибольшую суммарную прибыль интернет-магазину за указанный период?
orders_products['total_profit'] = orders_products['Количество'] * orders_products['Price']
mask = orders_products['Оплачен'] == 'Да'
orders_products[mask].groupby('ID Покупателя')['total_profit'].sum().sort_values(ascending=False)
```
